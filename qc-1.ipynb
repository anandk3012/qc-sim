## Basic Quantum Circuits and Statevector Representations

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import qiskit 
```

```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import  Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram, plot_bloch_vector, plot_state_qsphere
from qiskit_aer import AerSimulator
```

```python
qc = QuantumCircuit(QuantumRegister(3,'qubit'),ClassicalRegister(3,'cbit'))
qc.h(0)
qc.cx(0,1)
qc.x(2)
qc.draw('mpl')
```

```python
qc_vector = Statevector(qc)
qc_vector.draw('latex')
```

```python
plot_bloch_multivector(qc_vector)
```

```python
plot_state_qsphere(qc_vector)
```

```python
qc.measure(range(3),range(3))
qc.draw('mpl')
```

```python
backend = AerSimulator()
job = backend.run(qc, shots=1024, memory=True)
results = job.result()
counts = results.get_counts()
print(counts)
plot_histogram(counts)
```
